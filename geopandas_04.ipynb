```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')

# ဒီ code က google colab နဲ့ google drive ချိတ်ဖို့ သုံးတာပါ
# github ထဲမှာ mimu shp file တွေတင်ပေးထားပါတယ်ဗျ။
```

```python
township=gpd.read_file('/content/drive/MyDrive/Shape File/mmr_polbnda_250k_adm3_mimu.zip')
town=gpd.read_file('/content/drive/MyDrive/Shape File/mmr_pplp1_mimu_250k.zip')
road=gpd.read_file('/content/drive/MyDrive/Shape File/mmr_rdsl_250k_mimu.zip')
```

```python
# sample plotiing
township.plot()
```

```python
township.envelope.plot();
```

```python
# plotting multiple shp file

boundary=township.plot(figsize=(15,15),color='0.5')
road.plot(ax=boundary,color='red',alpha=0.3,label='road')
town.plot(ax=boundary,color='yellow',marker='*',markersize=3,label='town')

boundary.set_axis_off()
plt.legend(loc='upper left',frameon=False)
```

```python
# Filtering Data
township[township.TS=='Lahe'].plot()
```

```python
# filtering multiple data

township[(township.TS=='Katha')|(township.TS=='Indaw')|(township.TS=='Kawlin')].plot()
```
