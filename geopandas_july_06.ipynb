```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
township=gpd.read_file('/content/drive/MyDrive/Shape File/mmr_polbnda_250k_adm3_mimu.zip')
township.plot()
```

```python
township.plot(figsize=(10,10),cmap='magma')
```

```python
# boundary  plot 
township.boundary.plot(figsize=(10,10))
```

```python
# plotting multilple shapes
katha_dist=township[township.TS.isin(['Katha','Indaw','Tigyaing'])]
katha_dist.plot(figsize=(14,12),cmap='tab10')
```

```python
# outlines  label 

fig=plt.figure(1,figsize=(10,10))
ax=fig.add_subplot()

katha_dist.apply(lambda x: ax.annotate(s=x.TS,xy=x.geometry.centroid.coords[0],ha='center',fontsize=14),axis=1);
katha_dist.boundary.plot(ax=ax,color='black',linewidth=2)
katha_dist.plot(ax=ax,cmap='Pastel2',figsize=(12,12))

ax.text(-0.05,0.5,'super villain',transform=ax.transAxes,
        fontsize=20,color='gray',alpha=0.5,
        ha='center',va='center',rotation='90')
fig.tight_layout()
```
