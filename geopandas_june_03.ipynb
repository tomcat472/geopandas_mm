```python
# import libraries
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
world=gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities=gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
```

```python
# Base Map
base=world.plot()
cities.plot(ax=base,marker='*',color='k',markersize=2)
base.set_title('Sample Map');
```

```python
fig,ax=plt.subplots(figsize=(12,12))
world.plot(ax=ax)
cities.plot(markersize=3,color='red',ax=ax)
plt.tight_layout()
plt.show()
```

```python
# adding legend
fig,ax=plt.subplots(figsize=(10,8))
asia=world[world.continent=='Asia']
asia.plot(ax=ax,column='pop_est',legend=True)
```

```python
ax=world.plot(column='pop_est',legend=True)
cities.plot(ax=ax,color='red',markersize=2)
```
